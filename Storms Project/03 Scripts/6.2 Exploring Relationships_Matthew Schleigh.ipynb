```python
# Include matplotlib option for ease of calling charts

%matplotlib inline
```

```python
# Create a correlation matrix for heatmap

df_viz.corr()
```

```python
# Create a correlation heatmap with matplotlib

plt.matshow(df_viz.corr())
plt.show
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(10, 10)) # figure size 
plt.matshow(df_viz.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_viz.shape[1]), df_viz.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(df_viz.shape[1]), df_viz.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=12) # add font size
plt.title('Storms Correlation Matrix', fontsize=14) # add title
```

# Discuss what the coefficients in the plot mean in terms of the relationships between the variables.

# 6. Create a categorical plot and interpret the results.

### Let's examine hurricane_force_wind_diameter using a histogram

```python
df_viz_2['hurricane_force_wind_diameter'].value_counts(dropna=False).sort_index()
```

```python
# Create histogram

sns.histplot(df_viz_2['hurricane_force_wind_diameter'], bins = 30, kde=True)
```

### Such a high frequency of zeros is skewing the distribution. These will be omitted in creating the categories.

```python
# Create categories based on distribution

df_viz_2.loc[df_viz_2['hurricane_force_wind_diameter'] < 1, 'Hurricane size'] = 'No storm'
            
```

```python
df_viz_2.loc[(df_viz_2['hurricane_force_wind_diameter'] >= 1) & (df_viz_2['hurricane_force_wind_diameter'] < 100), 'Hurricane size'] = 'Small'
```

```python
df_viz_2.loc[(df_viz_2['hurricane_force_wind_diameter'] >= 100) & (df_viz_2['hurricane_force_wind_diameter'] < 200), 'Hurricane size'] = 'Medium'
```

```python
df_viz_2.loc[df_viz_2['hurricane_force_wind_diameter'] >= 200, 'Hurricane size'] = 'Large'
            
```

```python
# Check categories

df_viz_2['Hurricane size'].value_counts(dropna=False)
```

```python
# Drop 'No storm' observations

df_viz_2.drop(df_viz_2[df_viz_2['Hurricane size'] == 'No storm'].index, inplace=True)
```

```python
df_viz_2['Hurricane size'].value_counts(dropna=False)
```

```python
# Create a categorical plot in seaborn

sns.set(style='ticks')
g = sns.catplot(x='year', y='hurricane_force_wind_diameter', hue='Hurricane size', data=df_viz_2)
```

## Creating categories based on hurricane wind diameter size alone (small, medium, and large), and removing storms with a size of "0", shows that over time the frequency of storm size has largely remained static. A few of the largest storms have appeared only within the last few years, which should be noted. And the largest storm by wind diameter happened ten years ago, but looks to be an outlier based on its position along the plot.
