# Import libraries and data

```python
import pandas as pd
import numpy as np
import os

import geopandas as gpd
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

import matplotlib.pyplot as plt

import tqdm
from tqdm._tqdm_notebook import tqdm_notebook
```

```python
# Define path

path = r'/Users/matthewschleigh/Desktop/CareerFoundry/Achievement 6/Storms Project'
```

```python
# import data

df = pd.read_excel(os.path.join(path, '02 Data', 'Original Data', 'storms_US ONLY.xlsx'))
```

```python
df.shape
```

```python
df.head()
```

```python
locator = Nominatim(user_agent='myGeocoder')
```

```python
# create new column combining lat and lon

df['geom'] = df['latitude'].map(str) + ',' + df['longitude'].map(str)
```

```python
df.head()
```

```python
rgeocode = RateLimiter(locator.reverse, min_delay_seconds=0.001)
```

```python
df['address'] = df['geom'].apply(rgeocode)
df.head()
```

```python
df['address'].value_counts(dropna=False)
```

## Export new df

```python
df.to_csv(os.path.join(path, '02 Data', 'Prepared Data', 'storms_address.csv'))
```
